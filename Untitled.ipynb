```python
X=new_data[['LIMIT_BAL', 'SEX', 'EDUCATION', 'MARRIAGE', 'AGE','billnum', 'billamt']]
y=new_data[['nodebt', 'paidin15days', 'paidin1mnt', 'paidinqtr',
       'paidinhalfyear', 'paidinyearormore']]
# from sklearn.preprocessing import StandardScaler
# ss=StandardScaler()
# X_std=ss.fit_transform(X)
# from sklearn.preprocessing import PolynomialFeatures
# pf=PolynomialFeatures(3)
# X_Poly=pf.fit_transform(X_std)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.20, random_state=12)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.multioutput import MultiOutputClassifier
forest = RandomForestClassifier(n_estimators=20,random_state=0,max_features=2, max_depth=2)
multi_target_forest = MultiOutputClassifier(forest, n_jobs=-1)
multi_target_forest.fit(X_train, y_train)
```

```python
print("Accuracy score (training): {0:.3f}".format(multi_target_forest.score(X_train, y_train)))
print("Accuracy score (validation): {0:.3f}".format(multi_target_forest.score(X_test, y_test)))
```

```python
predictions = pd.DataFrame(multi_target_forest.predict(X_test),columns=y_test.columns)
# ['nodebt', 'paidin15days', 'paidin1mnt', 'paidinqtr', 'paidinhalfyear','paidinyearormore']
print("Confusion Matrix for nodebt:")
print(confusion_matrix(y_test.nodebt, predictions.nodebt))

print("Classification Report for nodebt:")
print(classification_report(y_test.nodebt, predictions.nodebt))

print("Confusion Matrix for paidin15days:")
print(confusion_matrix(y_test.paidin15days, predictions.paidin15days))

print("Classification Report for paidin15days:")
print(classification_report(y_test.paidin15days, predictions.paidin15days))

print("Confusion Matrix for paidin1mnt:")
print(confusion_matrix(y_test.paidin1mnt, predictions.paidin1mnt))

print("Classification Report for paidin1mnt:")
print(classification_report(y_test.paidin1mnt, predictions.paidin1mnt))

print("Confusion Matrix for paidinqtr:")
print(confusion_matrix(y_test.paidinqtr, predictions.paidinqtr))

print("Classification Report for paidinqtr:")
print(classification_report(y_test.paidinqtr, predictions.paidinqtr))

print("Confusion Matrix for paidinhalfyear:")
print(confusion_matrix(y_test.paidinhalfyear, predictions.paidinhalfyear))

print("Classification Report for paidinhalfyear:")
print(classification_report(y_test.paidinhalfyear, predictions.paidinhalfyear))

print("Confusion Matrix for paidinyearormore:")
print(confusion_matrix(y_test.paidinyearormore, predictions.paidinyearormore))

print("Classification Report for paidinyearormore:")
print(classification_report(y_test.paidinyearormore, predictions.paidinyearormore))
```

```python
from sklearn.model_selection import RandomizedSearchCV
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
print(random_grid)
```

```python
# Use the random grid to search for best hyperparameters
# First create the base model to tune
rf = RandomForestClassifier()
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations, and use all available cores
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 100, cv = 3, verbose=2, random_state=42, n_jobs = -1)
# Fit the random search model
rf_random.fit(X_train, y_train)
```
